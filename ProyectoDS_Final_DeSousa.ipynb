```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Leo el csv limpio resultado de la pre entrega
df = pd.read_csv('Emisiones_de_carbono_clean.csv')
df.describe().round(1)
```

```python
#Chequeo que realmente no me haya quedado ningún null despues de la limpieza
df.isna().sum()
```

En la entrega anterior ya habiamos hecho un análisis exploratorio, limpiado el dataset, reemplazado nulls numéricos con la media, eliminado algunas columnas y creado algunas columnas nuevas en base a otras.

**Usaremos a continuación algoritmos de Aprendizaje supervisado porque en este caso conocemos "Y" o nuestro valor a predecir, que será el cambio en la temperatura.**

```python
# Elimino columnas no numéricas (como area) o que pueden corromper mi modelo (año), solo quiero quedarme con datos de las emisiones de carbono y población

df_numeric = df.drop(['Area', 'Continent', 'Year'], axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

#Voy a evaluar importancia de features para reducir aún mas

X = df_numeric.drop('Average Temperature °C', axis=1)
y = df_numeric['Average Temperature °C']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el modelo Random Forest
forest = RandomForestRegressor(n_estimators=100, random_state=42)
forest.fit(X_train, y_train)

# Mostrar las características más importantes
importances = forest.feature_importances_
feature_importance = pd.Series(importances, index=X.columns).sort_values(ascending=False)
print(feature_importance)
```

Food Retail es la variable con mayor importancia en mi modelo de Random Forest, lo que indica que las emisiones relacionadas con la venta de alimentos tienen el mayor impacto en los cambios de temperatura.
Las variables relacionadas a incendios son los de menos importancia.

```python
# Elimino las columnas de menor importancia
columns_to_drop = ['Fires in organic soils', 'Fires in humid tropical forests', 'Savanna fires', 'Forest fires']
df_cleaned = df_numeric.drop(columns=columns_to_drop)
```

```python
# Voy a volver a ntrenar reajustando parámetros y con columnas de mayor importancia
from sklearn.model_selection import GridSearchCV

# Defino los hiperparámetros a probar
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestRegressor(random_state=42)

# Uso GridSearchCV para encontrar la mejor combinación de hiperparámetros
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Imprimo los mejores hiperparámetros encontrados
print("Mejores Hiperparámetros:", grid_search.best_params_)

# Entreno el modelo con los mejores hiperparámetros
best_rf = grid_search.best_estimator_
best_rf.fit(X_train, y_train)
```

```python
#Evaluación del modelo Random Forest
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

y_pred = best_rf.predict(X_test)
# Calcular RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")

# Calcular R² (Coeficiente de determinación)
r2 = r2_score(y_test, y_pred)
print(f"R²: {r2}")
```

Según la evaluación RMSE los resultados del modelo son razonables ya que la variación total de mi variable es de aproximadamente 5°C y un resultado de 0.39 representa un error relativamente bajo.

Los resultados según R2 no son prometedores en este modelo porque el resultado de la evaluación no se acerca a 1. Puede deberse a que no hay una relación lineal fuerte entre nuestros datos y la variabilidad es sútil.

**En el caso de estos datos creo que es mas útil ver el RMSE porque mi interés principal es predecir la temperatura con la mayor precisión posible, el RMSE nos da una medida clara de cuán cerca o lejos estamos en promedio de la temperatura real. Mientras más bajo sea, más precisas serán tus predicciones.**

```python
#Voy a entrenar a continuación un modelo SVR para el que si se necesitan normalizar los datos

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Dividir el dataset escalado en entrenamiento (70%) y prueba (30%)
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVR

# Crear un modelo SVR
svr_model = SVR(C=1.0, epsilon=0.1, kernel='rbf')
svr_model.fit(X_train_scaled, y_train)
```

```python
#Evaluación del modelo SVR
y_pred_svr = svr_model.predict(X_test_scaled)

# Calcular RMSE
rmse_svr = np.sqrt(mean_squared_error(y_test, y_pred_svr))
print(f"RMSE SVR: {rmse_svr}")

# Calcular R²
r2_svr = r2_score(y_test, y_pred_svr)
print(f"R² SVR: {r2_svr}")
```

Los resultados del modelo de SVR son peores que que los de Random Forest. Según R2 el desempeño es pésimo y según RMSE el error promedio en cambios de temperatura es mayor.

```python
#Ahora voy a probar un modelo de Gradient Boosting Regressor

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
gbr_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
gbr_model.fit(X_train_scaled, y_train)
```

```python
#Evaluación del modelo de Gradient Boosting
y_pred_gbr = gbr_model.predict(X_test_scaled)

# Calcular RMSE
rmse_gbr = np.sqrt(mean_squared_error(y_test, y_pred_gbr))
print(f"RMSE Gradient Boosting: {rmse_gbr}")

# Calcular R²
r2_gbr = r2_score(y_test, y_pred_gbr)
print(f"R² Gradient Boosting: {r2_gbr}")
```

Los resultados de la evaluación del modelo de Gradient Boosting nos dicen que de los 3 modelos, este no sería ni el mejor ni el peor.

Según R2 el modelo es mucho mejor que SVR y un poco peor que Random Forest aunque no por mucho.

El valor de RMSE nos indica lo mismo, su desempeño está casi en el nivel de Random Forest pero un poco por debajo
