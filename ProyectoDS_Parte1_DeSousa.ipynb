Este dataset describe las emisiones de CO2 relacionadas con el sector agroalimentario, las cuales representan aproximadamente el 62% de las emisiones anuales globales.
Me pareció interesante este tema para estudiar e intentar predecir cambios de temperatura, ya que comprender y abordar el impacto ambiental de la industria agroalimentaria es crucial para mitigar el cambio climático y desarrollar prácticas sostenibles dentro de este sector.

**Glosario**

*Las emisiones están representadas en kilotones (kt) y la temperatura en celsius.

**Columna: descripción (ejemplo del dato)**

*   Area: territorio geográfico (Afghanistan)
*   Year: año (2020)
*   Savanna fires: emisiones de incendios en sabanas (1.65)
*   Forest fires: emisiones de incendios en bosques (65)
*   Crop Residues: emisiones por quema o descomposición de residuos luego de la cosecha (104)
*   Rice Cultivation: emisiones de metano liberadas durante el cultivo de arroz (535)
*   Pesticides Manufacturing: emisiones por la producción de pesticidas (116)
*   Food Transport: emisiones por transporte de comida (205)
*   Food Household Consumption: emisiones por consumo de alimentos en el hogar (11.4)
*   Food Retail: emisiones relacionadas a la venta de alimentos (172)
*   On-farm Electricity Use: consumo eléctrico en granjas(500)
*   Food Packaging: emisiones por la producción y desecho de materiales para empaquetado de alimentos (282)
*   Agrifood Systems Waste Disposal: emisiones de la eliminación de residuos en el sistema agroalimentario (901)
*   Food Processing: Emisiones del procesamiento de productos alimenticios (345)
*   Fertilizers Manufacturing: Emisiones de la producción de fertilizantes (360)
*   IPPU: Emisiones de los procesos industriales y del uso de productos (804)
*   Manure applied to Soils: Emisiones por la aplicación de estiércol en suelos agrícolas (16.3)
*   Manure left on Pasture: Emisiones del estiércol en pastizales o tierras de pastoreo (140)
*   Manure Management: Emisiones del manejo y tratamiento del estiércol (270)
*   Fires in organic soils: Emisiones de incendios en suelos orgánicos (0)
*   Fires in humid tropical forests: Emisiones de incendios en bosques tropicales (9.58)
*   On-farm energy use: Consumo de energía en las explotaciones agrícolas (141)
*   Rural population: Número de personas que viven en áreas rurales (14591.0)
*   Urban population: Número de personas que viven en áreas urbanas (14591.0)
*   Total Population - Male: Número total de individuos masculinos en la población (14591.0)
*   total_emission: Emisiones totales de gases de efecto invernadero de varias fuentes (14591.0)
*   Average Temperature °C: El aumento anual promedio de la temperatura en grados Celsius (1.2825)

```python
#Reemplazar nulls numéricos con la media
numeric_columns = df.select_dtypes(include=['float64']).columns
null_columns = df[numeric_columns].columns[df[numeric_columns].isnull().any()]
for column in null_columns:
  df[column] = df[column].fillna(df[column].mean())
```

```python
#matplotlib - gráficos con valores normalizados para poder ver los patrones de cada variable
def normalizer(df):
    norm = (df - df.max()) / (df.max() - df.min())
    return norm

temp_emission = df.groupby("Year").agg({"Average Temperature °C": "mean", "total_emission": "mean", "Urban population": "mean", "Rural population": "mean"})
norm_temp = normalizer(temp_emission)


norm_temp.plot(figsize=(15, 4))
plt.title("Emisiones de CO2 & Cambios de Temperatura")
plt.show()
```

```python
#Matplotlib - Top 10 de mayor emisiones del úñtimo año (2020 es el mas reciente en el dataset)
def country_emission(df,year, length = 10):
        df = df.copy()
        plot = df.loc[df["Year"]==year]
        plot = plot.sort_values(by = "total_emission", ascending = True).tail(length)
        colors = plt.cm.get_cmap('Set2', len(plot))
        plt.figure(figsize=(8, 12))
        plt.barh(plot['Area'],
                plot['total_emission'],
                color=colors(range(len(plot))))
        plt.title(f'Emisiones de CO2 en el top {length} de territorios en {year}')
        plt.xlabel('Emisiones de CO2 en kilotoneladas')

        plt.show()

country_emission(df, year=2020)
```

```python
#matplotlib - Comparando emisiones de tipos de incendios distintos
columnas = ['Savanna fires', 'Forest fires', 'Fires in organic soils', 'Fires in humid tropical forests']
valores = df[columnas].mean()
plt.figure(figsize=(8, 6))
colores_fogosos = ['#8B0000','#FF4500','#4B4B4B','#FFA500']
plt.bar(columnas, valores, color=colores_fogosos)
plt.xlabel('Tipo de Incendios')
plt.ylabel('Emisiones')
plt.title('Emisiones por Tipo de Incendios')
plt.xticks(rotation=15)
plt.show()
```

```python
#Nueva columna de población total
df["Total_pop"] = df["Total Population - Male"] + df["Total Population - Female"]

#Me quedo solo con la población total y no con los géneros
df = df.drop(columns = ['Total Population - Male','Total Population - Female'],axis = 1)
```

```python
# Seaborn - Usando 3 variables: población, emisiones y temperatura
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df,
                          x="Average Temperature °C",
                          y="total_emission",
                          size="Total_pop",
                          hue="Continent",
                          palette="Set2",
                          sizes=(10, 60),
                          alpha=0.7
                         )

handles, labels = scatter.get_legend_handles_labels()
continent_handles = handles[:len(df["Continent"].unique())]
continent_labels = labels[:len(df["Continent"].unique())]

plt.legend(continent_handles, continent_labels, loc='upper right')

plt.title('Emisiones de CO2 y Temperatura por Población', fontsize=14, fontweight='bold')
plt.show()

colors = sns.color_palette("Set2")
```

```python
#Pairplot para explorar relación entre emisiones y población sin la temperatura
sns.pairplot(df, vars=["total_emission", "Urban population", "Rural population"], hue="Continent", palette="Set2")
plt.suptitle("Relación entre Emisiones y Población", y=1.02, fontsize=16)
plt.show()
```
